```python

"""
Base de @mariano
"""

import sympy as sp
from sympy.abc import s #importamos simbolo s
from IPython.display import display, Math

#Declaro simbolos variables
V1, V2, Vx = sp.symbols("V1, V2, Vx")
G1, G2, Y1, Y2, R3, C = sp.symbols("G1, G2, Y1, Y2, R3, C")

aa = sp.solve([ Vx*(Y1+Y2)-V1*Y1, 
                Vx*(G1+G2) -V1*G1 -V2*G2], 
                [V1, V2])

transf_func = aa[V2]/aa[V1]


# Ejercicio 7.a: Pasatodo de 1er orden

tf7a = transf_func.subs(Y1, s*C)
tf7a = tf7a.subs(Y2, 1/R3)


num, den = sp.fraction(sp.simplify(sp.expand(tf7a)))

num = sp.Poly(num,s)
den = sp.Poly(den,s)

#Obtengo coeficiente de mayor grado del Num y Den LC()
#Lo divido para obtener el coeficiente externo
k = num.LC() / den.LC()

num = num.monic()
den = den.monic()

tf7a_final = sp.Mul(k,num/den, evaluate=True)

print('')
print('################')
print('# Ejercicio 7a #')
print('################')
display(tf7a_final)
```

### Normalizacion


```python
tf7a = tf7a.subs(C, 1)
tf7a = tf7a.subs(R3, 1)

num, den = sp.fraction(sp.simplify(sp.expand(tf7a)))

num = sp.Poly(num,s)
den = sp.Poly(den,s)

#Obtengo coeficiente de mayor grado del Num y Den LC()
#Lo divido para obtener el coeficiente externo
k = num.LC() / den.LC()

num = num.monic()
den = den.monic()

tf7a_final = sp.Mul(k,num/den, evaluate=True)

print('')
print('################')
print('# Ejercicio 7a #')
print('################')
display(tf7a_final)
```

## 5.Utilidad
Estos circuitos que rotan la fase 180 se podria utilizar para los sistemas balanceados donde una se√±al se invierte (defasaje de 180) y luego se transmiten de manera diferencial. EL ruido que ingresa en la se√±al durante la transmision luego se va a cancelar cuando se resten estas dos se√±ales. Caso de ejemplo:
<img src="./balance_system.png" align="center" /> 

De todas maneras la aplicacion mas comun para estos circuitos es para corregir el error generado por el retardo de grupo en los medios de enlaces donde dependiendo de la frecuencia de cada se√±al se transmitira a distintas velocidades generando defasajes y distorsion en la se√±al de llegada. Este tipo de circuitos que tienen mucho mas control sobre la fase pueden ser usados para contrarestar.
